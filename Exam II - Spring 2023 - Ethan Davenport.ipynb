##### Words from full quotes. (This is given to you - you don't need to know regular expressions at the moment)

Here is a function that takes
a full quote (string) as argument and outputs a list of the words (all the words that were contained in the quote). The returned
words are all lower-case, and can contain only characters, digits, or
underscore. 

```python
import re 

def get_words_from_quote(full_quote):
    "This function takes a string quote and returns a list of all its words"
    return list(filter(None, re.split('[^a-z0-9_]+', full_quote.lower())))
```

Here is an example of the function in action:

```python
get_words_from_quote('I am only brave when I have to be!')
```

1) Open the processed_quotes.txt file and clean it up by removing the \n break into a new line character.

```python
# write your code here

f = open('processed_quotes.txt','r',encoding = 'utf-8')
lines = f.readlines()

lines = [line.strip('\n') for line in lines]
```

2) Build a postings-list dictionary.

A postings-list is
a dictionary whose **keys are full quotes**, and whose **values are themselves
dictionaries with key being a word, and value being the number of times
the word occurs in the full quote**. 

So, for the key “The heart has its reasons, of which the mind knows nothing. - Blaise Pascal”, the value will itself
be the following dictionary: {’the’:2, ’heart’:1, ’has’:1, ’its’:1,
’reasons’:1, ’of’:1, ’which’:1, ’mind’:1, ’knows’:1, ’nothing’:1,
’blaise’:1, ’pascal’:1}.

You may use the given function **get_words_from_quote** to break any quote in a list of words as you attempt to answer this question.

**Hint** (if it helps): You could first process the quote to a list of words and then make the dictionary with the word frequencies that will end up being the value for the key-quote of the final dictionary.  

If the hint disrupts your thoughts feel free to follow your own method.

```python
# write your code here

postings_list = {}
for line in lines:
    words = get_words_from_quote(line)
    # create the dictionary of words and number of times that words occurs
    D = {}
    for word in words:
        if word in D:
            D[word] += 1
        else:
            D[word] = 1
    # create the main dictionary with this dictionary as the key        
    postings_list[line] = D

postings_list
```

### Question 2 (25pts):

You are given a text document that contains words of an English Dictionary. Load and read the given text file and answer/do the following: 

1) Verify that you have words from all the letters of the english alphabet. DO NOT do this manually by checking each letter of the english alphabet, find a smarter/pythonic way of verifying that.(Hint: Can you get all unique first letters from the words and verify that they are 26?)   
2) Create a dictionary with keys all the letters of the english alphabet and values a list of all the words that start with that letter and are present in the given text file  
3) Create a dictionary with keys all the letters of the english alphabet and values the percentage (out of the total) of words that begin with each letter.    
4) Verify that the sum of all the percentages is **approximately** 100% (allowing for a small rounding error, it's ok if you do not get 100% in your answer).   


```python
# write your code here

f = open('English_Words.txt','r',encoding = 'utf-8')
lines = f.readlines()

letters = []
for line in lines:
    if line[0] not in letters:
        letters.append(line[0])

len(letters) == 26
```

```python
# part 2

D = {}
for letter in letters:
    letter_words = []
    for line in lines:
        if line[0] == letter:
            letter_words.append(line.strip('\n'))
    
    D[letter] = letter_words
D
```

```python
# part 3

total_count = len(lines)
D_pct = {}

for letter,words in D.items():
    letter_count = len(words)
    letter_pct = letter_count / total_count * 100
    D_pct[letter] = round(letter_pct,2)

D_pct
```

```python
# part 4

sum(D_pct.values())
```

### Question 3 (25pts):

Write a function that removes all vowels for a string (for this exercise y is not considered a vowel) and returns the vowel-les string. Spaces and puctuation are not affected. Include a docstring that explains how your function works. 

```python
#write your code here

def remove_vowels(string):
    '''
    The purpose of this function is to remove all vowels in a string.
    The function converts a string into a list so that it can iterate over that list, removing vowels along the way.
    The function will then convert the list back into a string, with vowels now removed.
    It expects a string input, and will return a string output.
    '''
    string = list(string)
    vowelless = [char for char in string if char.lower() not in 'aeiou']
    return ''.join(vowelless)

remove_vowels('Spaces and puctuation are not affected. Include a docstring that explains how your function works.')
```
